## １ページ目：物件のurlを取得

```
areas = []
area=13101
for i in range(23):
  areas.append(area)
  area+=1
```

```
top_page_url = 'https://www.chintai.net/tokyo/area/'
urls = []
for area in areas:
      urls.append(top_page_url+format(area)+'/list/')
print(urls)
```

```
#各地区の物件は多数のページに渡るのですべてのページのurlを取得
new_urls=[]

for url in urls:
    print(url)
    driver.get(url)
    new_urls.append(url)
    html = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(html, 'html.parser')
  
    body = soup.find('body')
    wrapper = body.find('div', {'id':'wrapper'})
    article= wrapper.find('div', {'id':'article'})
    section=article.find('div',{'class':"section"})
    list_pager=article.find('div',{'class':"list_pager"})
    
    p=list_pager.find('p') 
    total=p.find('span',{'class':"total"})
    total= str(total)
    total = total.split('>')[1].split('<')[0]
    total = re.findall(r'[0-9]', total)
    count=""
    for num in total:
      count+=num
      
    count=int(count)#地区の物件数が入る
    count=(count//30)+1 #ページ数に変換

    #各ページのurlを取得
    i=2
    while i<=count:
      new_url=url+'page{}/?o=7&o2=&prefkey=tokyo&rt=50'.format(i)
      new_urls.append(new_url)
      i+=1

    time.sleep(10)
```

```
links = []
#100ページ×30個の物件で十分

for url in new_urls[:100]:
    print(url)
    driver.get(url)
    html = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(html, 'html.parser')

    body = soup.find('body')
    wrapper = body.find('div', {'id':'wrapper'})
    article= wrapper.find('div', {'id':'article'})
    searchInputForm=article.find('form',{'id':"searchInputForm"})
    js_check=searchInputForm.find('section',{'class':"section js_check"})


    propertyBlocks=js_check.find_all('section',{'class':'mod_propertyBlock normal js_favorite_check_box gtm_aw_data'})

    for  propertyBlock in propertyBlocks:
      try:
        js_putRecent=propertyBlock.find('div',{'class':'pbCont js_putRecent js_block_link '})
        js_block_targets=js_putRecent.find_all('div')

        for js_block_target in js_block_targets:
          try:
            h3=js_block_target.find('h3')
            a=h3.find('a')
            link='https://www.chintai.net/'+a.get('data-detail-url')
            links.append(link)

          except:
            pass
    
      except:
        pass

    time.sleep(10)
```

```
link_df = pd.DataFrame()
link_df['link'] = links
link_df.head()
link_df.to_csv('link_csv/mansion_link.csv')
```

## ２ページ目： 物件ごとの処理

```
df = pd.read_csv('link_csv/mansion_link.csv')
df.head()
```

```
links = list(df['link'].unique())
```

```
len(links)
```

```
new_links = []
layouts = []

#最初に200個取ってくる
for link in links[:200]:
    print(link)
    driver.get(link)
    html = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(html, 'html.parser')

    try:
        body = soup.find('body')
        wrapper = body.find('div', {'id':'wrapper'})
        contents=wrapper.find("div",{'id':"l-contents"})
        article=contents.find("div",{'id':"article"})
        detailForm=article.find("form",{'name':"detailForm"})
        mt30=detailForm.find("section",{'class':'section mt30'})

        #画像を取得
        galleryArea=mt30.find("div",{'id':"detail_galleryArea"})
        gaLarge=galleryArea.find("div",{'id':"gaLarge"})
        gaLargeB=gaLarge.find("div",{'id':"gaLargeB"})
        carouselSet=gaLargeB.find("div",{'class':"carouselSet"})
        js_modal_btn=carouselSet.find("div",{'class':"target pic js_modal_btn"})

        ul=js_modal_btn.find('ul')
        li=ul.find('li')
        span=li.find('span')
        img=span.find('img')
        new_link = "http:"+img.get('src')
        new_links.append(new_link)
        temp_count = len(new_links)

        #間取りのデータ（１LDKなど）を取得
        #何故かピンポイントで取れないのでdiv全部とってきて探索する方式
        divs=mt30.find_all('div')
        for div in divs:
          try:
            table=div.find('table')
            spans=table.find_all('span',{'class':'bold'})

            for span in spans:
              try:
                 span=str(span)
                 span = span.split('>')[1].split('<')[0]
                 if 'K' in span:
                    layouts.append(span)
              except:
                 pass

          except:
            pass
    except:
       pass
    if len(layouts) != temp_count:
       layouts.append('OK')
       
    time.sleep(10)
```

```
#もう200個とる
for link in links[200:400]:
    print(link)
    driver.get(link)
    html = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(html, 'html.parser')

    try:
        body = soup.find('body')
        wrapper = body.find('div', {'id':'wrapper'})
        contents=wrapper.find("div",{'id':"l-contents"})
        article=contents.find("div",{'id':"article"})
        detailForm=article.find("form",{'name':"detailForm"})
        mt30=detailForm.find("section",{'class':'section mt30'})

        #画像を取得
        galleryArea=mt30.find("div",{'id':"detail_galleryArea"})
        gaLarge=galleryArea.find("div",{'id':"gaLarge"})
        gaLargeB=gaLarge.find("div",{'id':"gaLargeB"})
        carouselSet=gaLargeB.find("div",{'class':"carouselSet"})
        js_modal_btn=carouselSet.find("div",{'class':"target pic js_modal_btn"})

        ul=js_modal_btn.find('ul')
        li=ul.find('li')
        span=li.find('span')
        img=span.find('img')
        new_link = "http:"+img.get('src')
        new_links.append(new_link)
        temp_count = len(new_links)

        #間取りのデータ（１LDKなど）を取得
        divs=mt30.find_all('div')
        for div in divs:
          try:
            table=div.find('table')
            spans=table.find_all('span',{'class':'bold'})

            for span in spans:
              try:
                 span=str(span)
                 span = span.split('>')[1].split('<')[0]
                 if 'K' in span:
                    layouts.append(span)
              except:
                 pass


          except:
            pass
    except:
       pass
    if len(layouts) != temp_count:
       layouts.append('OK')
       
    time.sleep(10)
```

```
layouts[:10]
```

```
num_rooms = []
for layout in layouts:
    num_room_ = re.findall(r'[0-9]', layout)

    if not num_room_:
      num_rooms.append(0)
    else:
      for num_room in num_room_[0]:
          num_rooms.append(int(num_room))
```

```
link3_df = pd.DataFrame()
link3_df['link'] = new_links
link3_df['layout'] = layouts        
link3_df['num_room'] = num_rooms

link3_df = link3_df.sort_values(['num_room'], ascending=True)
link3_df.head()
```

## ３ページ目：取得したurlから画像をダウンロード

```
df = pd.read_csv('link_csv/pics_link.csv')
```

```
df.head()
```

```
for i in range(5):
    links = list(df[df['num_room']==i]['link'].unique())
    #条件抽出↑
    count = 0
    for pict_link in links:
      #画像ファイルのダウンロード
      r = requests.get(pict_link)
      count = count+1
      f = open('original_data/{}_room/{}room_{}.jpg'.format(i, i, count), 'wb')
      f.write(r.content)
      f.close()

      time.sleep(10)
```

```
#room_1の160までダウンロードしたらエラーで止まったのでそれ以降をやる
links = list(df[df['num_room']==1]['link'].unique())
count = 160
for pict_link in links[160:]:
  try:
      #画像ファイルのダウンロード
      r = requests.get(pict_link)
      count = count+1
      f = open('original_data/{}_room/{}room_{}.jpg'.format(i, i, count), 'wb')
      f.write(r.content)
      f.close()
      print(pict_link)
  except:
    pass  
  time.sleep(10)
```

```
#残りのダウンロード
i=2
while i<5:
    links = list(df[df['num_room']==i]['link'].unique())
    count = 0
    for pict_link in links:
      try:
          #画像ファイルのダウンロード
          r = requests.get(pict_link)
          count = count+1
          f = open('original_data/{}_room/{}room_{}.jpg'.format(i, i, count), 'wb')
          f.write(r.content)
          f.close()
      except:
        pass 
      time.sleep(10)
    i+=1
```
